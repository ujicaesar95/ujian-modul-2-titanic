### 8. Missing Values
#### Berapa banyak NaN age data dan mean dari setiap who?
#### Isi data NaN dari kolom age dengan mean dari umur tergantung whonya dia apa (child/woman/man)
#### Kemudian print banyak NaN age data dan mean umur setiap who lagi.

```python
print('Banyak NaN man & mean'+': '+str(sum(titanic[titanic['who']=='man']['age'].isna()))+' & '+str(titanic[titanic['who']=='man']['age'].mean()))
print('Banyak NaN woman & mean'+': '+str(sum(titanic[titanic['who']=='woman']['age'].isna()))+' & '+str(titanic[titanic['who']=='woman']['age'].mean()))
print('Banyak NaN child & mean'+': '+str(sum(titanic[titanic['who']=='child']['age'].isna()))+' & '+str(titanic[titanic['who']=='child']['age'].mean()))
```

```python
rata2=dict(man=titanic[titanic['who']=='man']['age'].mean(),woman=titanic[titanic['who']=='woman']['age'].mean(),child=titanic[titanic['who']=='child']['age'].mean())
titanic['age'].fillna(value=titanic['who'].apply(lambda x: rata2.get(x)),inplace=True)

print('Banyak NaN man & mean'+': '+str(sum(titanic[titanic['who']=='man']['age'].isna()))+' & '+str(titanic[titanic['who']=='man']['age'].mean()))
print('Banyak NaN woman & mean'+': '+str(sum(titanic[titanic['who']=='woman']['age'].isna()))+' & '+str(titanic[titanic['who']=='woman']['age'].mean()))
print('Banyak NaN child & mean'+': '+str(sum(titanic[titanic['who']=='child']['age'].isna()))+' & '+str(titanic[titanic['who']=='child']['age'].mean()))
```

### 9. Adding New Features/Columns/Variables

#### Buat DataFrame baru dengan column2 sebagai berikut
 - __id__ = id table ini (PK)
 - __mean_fare__ = Mean Fare
 - __std_fare__ = Std Fare
 - __batas_min_fare__ = Mean Fare - (3 x Std Fare)
 - __batas_max_fare__ = Mean Fare + (3 x Std Fare)
 
#### Dan data2nya ada 3 (fare setiap class yaitu first, second, and third class)

```python
unik=titanic['class'].unique()
TitanicOutCalc=pd.DataFrame(columns='id mean_fare std_fare batas_min_fare batas_max_fare'.split(),
                            data=[[x,titanic[titanic['class']==unik[x-1]]['fare'].mean(),titanic[titanic['class']==unik[x-1]]['fare'].std(), 
                                    titanic[titanic['class']==unik[x-1]]['fare'].mean()-(3*titanic[titanic['class']==unik[x-1]]['fare'].std()),
                                    titanic[titanic['class']==unik[x-1]]['fare'].mean()+(3*titanic[titanic['class']==unik[x-1]]['fare'].std())
                                    ] for x in range (1,len(unik)+1)])
TitanicOutCalc
```

#### Print outlier % of each class

```python
data={}
for index in range(len(unik)):
    data[unik[index]]=100*sum((titanic[titanic['class']==unik[index]]['fare']>TitanicOutCalc['batas_max_fare'][index])|(titanic[titanic['class']==unik[index]]['fare']<TitanicOutCalc['batas_min_fare'][index]))/len(titanic[titanic['class']==unik[index]])
    

print('Outlier % First' +': '+str(data['First']))
print('Outlier % Second' +': '+str(data['Second']))
print('Outlier % Third' +': '+str(data['Third']))
```

#### Tambahkan column outliercalcId (FK) di dfTitanic

```python
titanic['outliercalcId']=np.nan
FK=dict(Third=3,First=1,Second=2)
titanic['outliercalcId'].fillna(value=titanic['class'].apply(lambda x: FK.get(x)),inplace=True)
titanic.head()
```

```python
titanic['outliercalcId'].value_counts()
```

#### Tambahkan column outlier (1 for yes and 0 for no)

```python
titanic['outlier']=np.nan
out={True:1,False:0}
dataout=[None]*len(titanic)
for index in range(len(titanic)):
    dataout[index]=out[(titanic['fare'][index]>TitanicOutCalc['batas_max_fare'][titanic['outliercalcId'][index]-1])| (titanic['fare'][index]<TitanicOutCalc['batas_min_fare'][titanic['outliercalcId'][index]-1])]
titanic['outlier']=dataout
titanic['outlier'].value_counts()
```
